```python
# Importando as bibliotecas necessárias e escolhendo estilo dos gráficos
import pandas as pd # tratamento de dataframes
import matplotlib.pyplot as plt # criação de gráficos
plt.style.use("seaborn") # estilo dos gráficos
```

```python
# Criando o dataframe
df = pd.read_excel("AdventureWorks.xlsx")
```

```python
# Visualizando as primeiras linhas do dataframe
df.head()
```

```python
# Quantidade de linhas e colunas do dataframe:
df.shape
```

```python
# Entendendo sobre os tipos de dados do dataframe:
df.info()
```

```python
df.describe()
```

```python
# Valor total da Receita (Total de vendas):
round(df["Valor Venda"].sum(), 2)
```

```python
# Qual é o custo total?
# Devemos multiplicar o custo unitário pela quantidade de produtos

df['Custo total'] = df['Custo Unitário'] * df['Quantidade'] # Criando a coluna custo total
round(df['Custo total'].sum(), 2) # somando o valor da coluna custo total
```

```python
# Qual é o valor do lucro?
# Vamos subtrair o custo total da receita e obter o valor do lucro

df['Lucro'] = df['Valor Venda'] - df['Custo total']
round(df['Lucro'].sum(),2)
```

```python
# Visualizando o dataframe atual
df.head(1)
```

```python
# Quanto tempo leva para enviar os produtos?

df['Tempo_envio'] = df['Data Envio'] - df['Data Venda']
```

```python
df.head(2)
```

```python
# Tempo médio de envio por produto
# Será necessário transformar a coluna tempo envio em numérica para fazer a média

df['Tempo_envio'] = (df['Data Envio'] - df['Data Venda']).dt.days
```

```python
df.head(1)
```

```python
df.Tempo_envio.dtype
```

```python
# Média de tempo de envio por marca
round(df.groupby('Marca')['Tempo_envio'].mean(),1)
```

```python
# Verificando se existem valaores ausentes
df.isnull().sum()
```

```python
# Lucro por marca

df.groupby([df['Data Venda'].dt.year, "Marca"])['Lucro'].sum()
```

```python
# Ajustando o formato dos campos float
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
# Lucro por marca

df.groupby([df['Data Venda'].dt.year, "Marca"])['Lucro'].sum()
```

```python
# Criando um novo dataframe de um conjunto de dados resetando o index
lucro_ano = df.groupby([df['Data Venda'].dt.year, "Marca"])['Lucro'].sum().reset_index()
lucro_ano
```

```python
# Total de Produtos vendidos (Gráfico de barras horizontais)
df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=False)
```

```python
# Total de Produtos vendidos (Gráfico de barras horizontais)
df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=True).plot.barh(title='Total de produtos vendidos')
plt.xlabel('Total')
plt.ylabel('Produto');
```

```python
# Total de Lucro por ano (Gráfico de Barras)
df.groupby(df['Data Venda'].dt.year)['Lucro'].sum().sort_values(ascending=False).plot.bar(title='Total de Lucro por ano')
plt.xlabel('Ano da Venda')
plt.ylabel('Lucro');
```

```python
df.groupby(df['Data Venda'].dt.year)['Lucro'].sum()
```

```python
# Analisando apenas as vendas de 2009

filtro = df['Data Venda'].dt.year == 2009
df_2009 = df.loc[filtro]
```

```python
df_2009.head(2)
```

```python
df_2009.info()
```

```python
# Analisando o lucro por mês das vendas do ano de 2009

df_2009.groupby(df_2009["Data Venda"].dt.month)["Lucro"].sum().plot(title="Lucro x Mês")
plt.xlabel("Mês")
plt.ylabel("Lucro");
```

```python
# Avaliação do lucro por marcas:

df_2009.groupby("Marca")["Lucro"].sum().plot.bar(title="Lucro x Marca")
plt.xlabel("Marca")
plt.ylabel("Lucro")
plt.xticks(rotation='horizontal');
```

```python
df_2009.groupby("Classe")["Lucro"].sum().plot.bar(title="Lucro x Classe")
plt.xlabel("Classe")
plt.ylabel("Lucro")
plt.xticks(rotation='horizontal');
```
